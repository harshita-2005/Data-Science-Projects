```python
 #creating a dataframe
import pandas as pd
df=pd.read_csv('/content/Housing.csv')
df
```

```python
df.info()
```

```python
df=df.drop(['prefarea'],axis=1)
```

```python
df['mainroad'].replace(['yes','no'],['1','0'],inplace=True)
df['guestroom'].replace(['yes','no'],['1','0'],inplace=True)
df['basement'].replace(['yes','no'],['1','0'],inplace=True)
df['hotwaterheating'].replace(['yes','no'],['1','0'],inplace=True)
df['airconditioning'].replace(['yes','no'],['1','0'],inplace=True)

df['furnishingstatus'].replace(['furnished','unfurnished','semi-furnished'],['1','0','0.5'],inplace=True)
```

```python
df
```

```python
#visualization
import matplotlib.pyplot as plt
plt.plot(df['price'],df['area'])
```

```python
#dividing the data into i/p and o/p
x=df.iloc[:,1:].values
x
```

```python
y=df.iloc[:,0].values
y
```

```python
#tain_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
#normalization-only for i/p values
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
#apply a regressor
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#fit the model
model.fit(x_train,y_train)
```

```python
#predict the o/p

y_pred=model.predict(x_test)
y_pred
```

```python
import pandas as pd
df=pd.read_csv('/content/country_wise_latest.csv')
df
```

```python
df.info()
```

```python
#input- Confirmed,	Deaths,	Recovered,	Active,New cases,	New deaths,	New recovered	cols
```

```python
#divide the data into i/p
x=df.iloc[:,1:8].values
x
```

```python
#visualization
import matplotlib.pyplot as plt
plt.scatter(df['Confirmed'],df['Deaths'],c='r')
```

```python
#find no.of clusters
import numpy as np
np.sqrt(187)#187 is no of points/rows
#k= no of clusters
#k value should not exceed sqrt of total points
#hence k value should be in range of 2 and 13
```

```python
#1.ELBOW METHOD
#2.SILHOUETTE SCORE METHOD
from sklearn.cluster import KMeans
k=range(2,14) #my range is from
```

```python
#we will consider the point at which the elbow is more prominent
#we will consider k as 4 for now,but not sure
```

```python
#2. SILHOUETTE SCORE METHOD
from sklearn.metrics import silhouette_score
k=range(2,14)
for i in k:
  model=KMeans(n_clusters=i,random_state=0)
  model.fit(x)
  y_pred=model.predict(x)
  print(f'{i} clusters , score={silhouette_score(x,y_pred)}')
  plt.bar(i,silhouette_score(x,y_pred))
```

```python
#confirmation: the no.of clustera to be considered is 1.
```

```python
#apply clusterer
k=1
from sklearn.cluster import KMeans
model_f=KMeans(n_clusters=k,random_state=0)
model_f.fit(x)
```

```python
y=model_f.predict(x)#predicted o/p
y
```

```python
#this means all the points are in cluster no 1
```

```python
x[y==0,0]
#so the first '0' os cluster no 1 and the second '1' is col index one
#the value of input when cluster 1 is selected and clo index 1 is selected
```

```python
np.unique(y,return_counts=True)
```

```python
#FINAL VISUALIZATION
plt.figure(figsize=(10,5))
for i in range(k):
  plt.scatter(x[y==i,0],x[y==i,1],label=f'cluster {i}')
plt.scatter(model_f.cluster_centers_[:,0],model_f.cluster_centers_[:,1],100,c='yellow',label='centroids')
plt.legend()
```
